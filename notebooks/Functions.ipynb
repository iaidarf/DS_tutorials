**Непрерывность**. У непрерывной функции бесконечное число точек и для каждой точки есть свое конкретное значение. Непрерывность определяется с помощью понятия предела функции. Функция непрерывна в точке а, если $\lim\limits_{x \to a} f(x) = f(a)$ \



Если функция не непрерывна, значит есть разрыв (устранимый, "скачок", бесконечный)

**Гладкая функция** или **непрерывно дифференцируемая функция** — это функция, имеющая непрерывную производную на всём множестве определения.

**Монотонность функции** \
Монотонная функция — это функция, которая всё время либо возрастает, либо убывает.\
\
Функция y=f(x) называется строго **возрастающей** на промежутке, если большему значению аргумента из этого промежутка соответствует большее значение функции, т.е.\
f(x)↑: x1<x2 ⇒ f(x1)<f(x2) \
\
Функция y=f(x) называется строго **убывающей** на промежутке, если большему значению аргумента из этого промежутка соответствует меньшее значение функции, т.е.\
f(x)↓: x1<x2 ⇒ f(x1)>f(x2)

**Унимодальная функция** - функция у которой локальный экстремум является одновременно и глобальным.

## Производные функций
Производная - это мгновенная скорость изменения значения функции в заданной точке.

$$f'(x) = \lim_{\Delta x \to 0} \frac{f(x+\Delta x)-f(x)}{\Delta x}$$

На практике используются приближенные методы вычисления производных из-за сложности самих функций.

Производная функции тоже является функцией, поэтому для нее также актуальны те же самые свойства.


#### Связь произодной функции и **свойств** исходной функции:
Производная функция является **непрерывной**  - исходная функция является **гладкой**.\
Производная **положительная** на каком-то отрезке - исходная функция на этом отрезке **возрастает**.\
Производная **отрицательная** на каком-то отрезке - исходная функция на этом отрезке **убывает**.

Производные в DS. Используются для оптимизации целевых функций при обучении моделей. Задача оптимизации заключается в поиске максимального или минимального значения функции на заданном интервале (**экстремумы функции**). Экстремумы бывают локальные и глобальные.

### Экстремум функции.
В точках экстремумов производная равна нулю $f'(x_0)=0 $, если она определена. Это **необходимое**, но недостаточное условие. \
То есть, если в какой-то точке производная равна нулю, то она эта точка не обязательно является экстремумом.\
Также производная функции может быть неопределена в точках локальных экстремумов.

Экстремум:
extremum.JPG

Не экстремум:
not extremum.JPG

**Как отличить локальный минимум от локального максимума?** Если производная в точке равна нулю, то это говорит о том, что данная точка яляется экстремумом, но не определяет максимум этот или минимум.\
Для этого используется вторая производная.\
Если вторая производная положительная, значит первая производная возрастает. И если вторая производная отрицательная, значит первая производная убывает.\
\
Отсюда **достаточное условие экстремума**:\
$f'(x)=0$ и  $f''(x)>0$ - минимум \
$f'(x)=0$ и  $f''(x)<0$ - максимум

**Пример визуализации функции и производных**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-100,100)
plt.figure(figsize=(6,12))

# функция
plt.subplot(3,1,1)
plt.plot(x, x**2)

# первая производня
plt.subplot(3,1,2)
plt.plot(x, 2*x)
plt.hlines(y=0,xmin=-100, xmax=100, linestyle='--', color='r')

# вторая производная
plt.subplot(3,1,3)
plt.hlines(y=6, xmin=-100, xmax=100)
```

### Вычисление пределов

```python
import sympy as sym
```

```python
# x - символьная переменная
x = sym.Symbol('x')
```

```python
# лимит при х стремящемуся к нулю
sym.limit(sym.sin(x)/x,x,0)
```

```python
# лимит при х стремящемуся к бесконечности
sym.limit(sym.sin(x)/x,x,sym.oo)
```

```python
# лимит при х стремящемуся к нулю
sym.limit(1/x,x,0)
```

### Вычисление производных

```python
from sympy import diff, symbols, cos, sin
```

image.png

```python
# x - символьная переменная
x = symbols('x')
# функция
f = x**5 + 4*sin(2*x) + cos(3*x + 3)
f
```

```python
# продиффиринцируем эту функцию по x
d1 = diff(f, x)
d1
```

```python
# значение производной в точке x = 1
d1.subs(x, 1)
```

```python
# можно вычислить числовое выражение
d1.evalf(subs={x:1})
```

```python
# вторая производная
d2 = diff(f, x, x)
d2
```

# Линейные уравнения

Где применяется:
- задачи оптимизации
- задачи линейной регрессии
- аппроксимация функций
- интерполяция пропущенных значений и многое другое

```python
# решение системы уравнений
# 4x + 2y + z = 4
# x + 3y = 12
# 5y + 4z = -3
```

```python
import numpy as np
from numpy import linalg
a = np.array( [ [4, 2, 1], [1, 3, 0], [0, 5, 4] ] )
b = np.array([4, 12, 3])
c = linalg.solve(a, b)
print(f'x = {c[0]}, y = {c[1]}, z = {c[2]}' )
```
