# Уменьшение размерности входных данных

В анализе данных иногда бывает необходимо создать упрощенную модель, максимально точно описывающую реальное положение дел.

## Метод главных компонент
PCA - principal component analysis\
Данный метод аппроксимирует n-размерное облако наблюдений до эллипсоида (тоже n-мерного), полуоси которого и будут являться будущими главными компонентами. И при проекции на такие оси (снижении размерности) сохраняется наибольшее количество информации.


```python
import numpy as np
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
# тестовые данные
X = np.array([[-1,-1], [-2,-1], [-3,-2], [1,1], [2,1], [3,2]])
```

```python
# визуализируем
sns.scatterplot(data = X, x=X[:,0], y=X[:,1])
```

В данной выборке у нас имеются два признака, сильно коррелирующие друг с другом. С помощью алгоритма PCA мы сможем легко найти признак-комбинацию и, ценой части информации, выразить оба этих признака одним новым.

```python
# Параметр n_components указывает на количество измерений, на которые будет производиться проекция,
# то есть до скольки измерений мы хотим снизить наш датасет. 
pca = PCA(n_components = 1)
pca.fit_transform(X)
```

Проверим и прочие параметры модели – функция имеет ряд атрибутов, позволяющих получить доступ к промежуточным переменным:

```python
print ('Mean vector: ', pca.mean_)
print ('Projection: ', pca.components_)
print ('Explained variance ratio: ', pca.explained_variance_ratio_) # разбиение вариаций по главным компонентам
print ('Singular values', pca.singular_values_) # сингулярные значения для главных компонент
```

```python
sns.scatterplot(data = X, x=X[:,0], y=X[:,1])
sns.scatterplot(x=[pca.mean_[0]], y=[int(pca.mean_[1])])
```
