# Корреляция и регрессия

### Ковариация 

(ко - совместная, вариация - изменчивость). Мера **линейной** зависимости двух случайных величин.

Если ковариация положительна, то с ростом значений одной случайной величины, значения второй имеют тенденцию возрастать, а если знак отрицательный — то убывать.

$$ cov(X, Y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{x})}}{N - 1} $$
где N - количество случайных величин, а единица - количество степеней свободы.

Однако только по абсолютному значению ковариации нельзя судить о том, насколько сильно величины взаимосвязаны, так как масштаб ковариации зависит от их дисперсий. Значение ковариации можно нормировать, поделив её на произведение среднеквадратических отклонений (квадратных корней из дисперсий) случайных величин. Полученная величина называется **коэффициентом корреляции Пирсона**, который всегда находится в интервале от −1 до 1:

$$ r(x, y) = \frac{cov(x, y)}{\sigma_x\sigma_y}$$


### Подробнее про формулу корреляции

Давайте остановимся на формуле коэффициента корреляции, которую мы получили:
$$ r(x, y) = \frac{cov(x, y)}{\sigma_x\sigma_y}$$
запишем формулу чуть подробнее и выполним возможные преобразования:

$$ r(x, y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{(N - 1)\sqrt{\sum{\frac{(x_i - \bar{x})^2}{N-1}}}\sqrt{\sum{\frac{(y_i - \bar{y})^2}{N-1}}}} $$

теперь вынесем 1/ (N - 1) из под корней 

$$ r(x, y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{(N - 1)\frac{1}{(N-1)}\sqrt{\sum{(x_i - \bar{x})^2}}\sqrt{\sum{(y_i - \bar{y})^2}}} $$

и сократим (N - 1)

$$ r(x, y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{\sqrt{\sum{(x_i - \bar{x})^2}}\sqrt{\sum{(y_i - \bar{y})^2}}} $$

таким образом, мы сократили N - 1 в знаменателе и получили финальную формулу для коэффициента корреляции, которую вы часто сможете встретить в учебниках:

$$ r(x, y) = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{\sqrt{\sum{(x_i - \bar{x})^2}\sum{(y_i - \bar{y})^2}}} $$

**Корреляция не подразумевает причинно-следственных связей!**

```python
'''Демонстрация работы ковариации и корреляции'''
import numpy as np
import random as r
import matplotlib.pyplot as plt

def cov(x, y):
    assert x.size == y.size
    return ((x - x.mean()) * (y - y.mean())).sum()/(x.size - 1)

def cor(x, y):
    return cov(x, y)/(np.std(x, ddof=1)*np.std(y, ddof=1))

# функция имитирущая случаные факторы
# р - настолько существенным будет случайный фактор
def randomize(arr, p):
    alpha = np.max(arr) - np.min(arr)
    res = np.zeros(arr.shape)
    for i, v in enumerate(arr):
        sign = 1 if r.choice([True, False]) else -1
        res[i] = v + sign*alpha*r.random()*p
    return res


x = np.array(range(30))
y = randomize(x, 0.1)
y1 = randomize(x, 0.5)
y2 = randomize(x, 1)


fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 3))
ax1.scatter(x, y)
ax2.scatter(x, y1)
ax3.scatter(x, y2)
ax1.set_title('высокая корреляция')
ax2.set_title('средняя корреляция')
ax3.set_title('низкая корреляция')
plt.show()
```

```python
print(f'''
cov1: {cov(x, y):.2f}
cov2: {cov(x, y1):.2f}
cov3: {cov(x, y2):.2f}

cor1: {cor(x, y):.2f}
cor2: {cor(x, y1):.2f}
cor3: {cor(x, y2):.2f}
''')
```

#### Ограничения коэффициента корреляции Пирсона
1. Характер взаимосвязи должен быть линейный и монотонный. Для этого нужно построить диаграмму рассеяния и убедиться что характер  связи линейный.\
Примеры. Коэф. Пирсона показывает что взаимосвязи нет, хотя это не так. Просто она носит нелинейный характер.


2. Желательно, чтобы данные имели нормальное распределение.\
Пример с выбросом.


На такие случаи есть другие **непараметрические** коэффициенты корреляции:
- Спирмена
- тау Кенделла

Особенность в том, что при расчетах берутся не значения параметров, а их ранги

## Одномерный регрессионный анализ
Позволяет проверять гипотезы о взаимосвязи одной  количественной зависимой переменной и нескольких независимых.

### Простая линейная регрессия
Взаимосвязь двух переменных

Линейная регрессия (англ. Linear regression) — используемая в статистике регрессионная модель зависимости одной (объясняемой, зависимой) переменной $y$ от другой или нескольких других переменных (факторов, регрессоров, независимых переменных) $x$ с **линейной функцией зависимости**.

В общем виде функция линейной регрессии выглядит как:

$$ y = b_0 + b_1x $$
$b_0$ - intercept значение пересечения линии с осью Y 

$b_1$ - slope задаёт наклон линии регрессии

Строят регрессионную прямую **методом наименьших квадратов** (МНК)

МНК - это способ нахождения оптимальных параметров линейной регресссии ($b_0$,  $b_1$), таких, что сумма квадратов ошибок (остатков) была минимальная.\
Остаток - это расстояние от реального значения зависимой переменной до линии регрессии: $y - y_{pred} $ . Он может быть как положительным, так и отрицательным. Такой остаток считается для каждого значения. Далее квадраты остатков суммируются. И задача состоит в том, чтобы подобрать такую линию, чтобы сумма квадратов была минимальной.

$C(a,b) = (y_1 - y_{pred1})^2 + (y_2 - y_{pred2})^2 + .... + (y_n - y_{predn})^2$ 

Чем меньше значение этой ошибки, тем лучше модель. Оптимальные значение a и b - это те, при которых значение С достигает своего минимума.

Расчёт параметров идёт по таким формулам (приведена формула без вывода):

$$ b_1 = \frac{sd_y}{sd_x}r_{xy} $$
$$ b_0 = \bar{Y} - b_1\bar{X} $$
$sd$ - стандартное отклонение\
$r_{xy}$ - коэф. корреляции\
$\bar{Y}$ - среднее значение зависимой переменной\
$\bar{X}$ - среднее значение независимой переменной

```python
'''Демонстрация МНК'''
b1 = y1.std()/x.std()*cor(x, y1)
b0 = y1.mean() - b1*x.mean()
f = lambda x: b0 + b1*x
y_pred = f(x)
plt.scatter(x, y1)
plt.plot(x, y_pred, color='r')
plt.show()
```

### Коэффициент детерминации

Коэффициент детерминации $R^2$  равняется квадрату коэффициента корреляции $R^2 = r_{xy}^2$\
$R^2$ показывает в какой степени дисперсия одной переменной обусловлена "влиянием" другой переменной.

Вычисляется по формуле:

$$R^2 = 1 - SS_{res}/SS_{tot}$$ 
где:\
$SS_{res} = \sum_{i=0}^n(y_i - y_{predi})^2 $ - сумма квадратов ошибок модели ( Sum of Squared Error) \
$SS_{tot} = \sum_{i=0}^n(y_i - \bar y)^2 $ - сумма квадратов отклонений фактических значений от средней величины


$R^2$ может выступать индикатором того, насколько модель адекватна. $R^2$ принимает значения от 0 до 1. 

Если он равен 0, это означает, что связь между переменными регрессионной модели отсутствует и входная переменная плохо «объясняет» поведение выходной, т.е. линейная зависимость между ними отсутствует. 

Напротив, если коэффициент детерминации равен 1, это соответствует идеальной модели, когда все точки наблюдений лежат точно на линии регрессии, т.е. сумма квадратов их отклонений равна 0. Тут стоит задуматься - возможно модель "переобучилась".

Стоит ориентироваться на диапазоном значений 0.5...0.9, это будет говорить о том, что регресиия хорошо объясняет зависимость данных.


### Условия применения линейной регрессии с одним предиктором
- Линейная связь X и Y
- Нормальное распределение остатков
- Гомоскедастичность - постоянная изменчивость остатковна всех уровнях независимой переменной

Примереы по ссылке: https://gallery.shinyapps.io/slr_diag/

- гистограмму распределения остатков и QQ-plot(убедится в нормальности)


### Применение регрессионного анализа и интерпретация результатов

#### Пример

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data/states.csv')
df.head()
```

Есть данные по штатам с различными значениями:
 - **metro_res** - процент населения живущие в столице
 - **white** - процент белого населения
 - **hs_grad** - процент людей со образованием
 - **poverty** - уровень бедности
 - **female_house** - процент домов, где есть домохозяйки 
 
Исследуем связь уровня образования и бедности, где бедность будет ЗП, а уровень образования НП.

Первое, что нам необходимо сделать, это построить линейную модель, которая наилучшим образом будет описывать наши данные.

$$ \hat{y} = b_0 + b_1x $$

Далее, построив нашу модель, нам надо узнать, насколько хорошо наша объясняет ЗП, для этого найдём коэфицент детерминации $R^2$

Проверим нулевую гипотезу:
$$ b_1 = 0 : H0$$

Третья наша задача, это задача предсказания, по данным НП мы хотим предсказать ЗП.

```python
# Посмотрим на наши данные
sns.jointplot(x='hs_grad', y='poverty', data=df, kind='reg', color='m')
```

```python
df_descr = df.describe().transpose()
df_descr
```

```python
'''Построим модель'''
from scipy.stats import linregress

slope, intercept, r, p, std_err =  linregress(df['hs_grad'], df['poverty'])

x = np.linspace(75, 100)

reg = lambda x: intercept + slope*x
plt.scatter(x='hs_grad', y='poverty', data=df, label='data')
plt.xlabel('hs_grad %')
plt.ylabel('poverty %')
plt.title('Linear Regression')
plt.plot(x, reg(x), color='r', label='fitted line')
plt.legend()
plt.show()
```

```python
print(f'''
slope = {slope:.2f}
intercept = {intercept:.2f}
r = {r:.2f}
r squared = {(r ** 2):.2f}
p = {p:.5f}
std_err = {std_err:.3f}
''')
```

**Интерпретация результатов** \
Уравнение принимает вид: $y_{pred} = b_0 + b_1x = 64,78 -0,62x$ \
Смысл коэф. $b_1$ перед независимой переменной: он показывает насколько будет изменяться зависимая переменная при изменении независимого на одну единицу. \
Уровень значимости гораздо меньше значение 0.05, что позволяет отклонить нулевую гипотезу, то есть между двумя переменными существует статистически значимая связь.\
Коэф. $R^2 = 0.56$ говорит о том, что 56% изменчивости зависимое переменной объясняется полученной моделью. Это не так уж и много. Получается что чуть меньше половины изменчивости обуславливается не включенными в модель факторами.

## Корреляция и регрессия на примере boston

Общий подход к решению:
+ Определение формы зависимости
+ Построение модели регрессии
+ Оценка неизвестных значений зависимой переменной

```python
import pandas as pd
from sklearn.datasets import load_boston
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
# нужный датасет встроен в библиотеку sklearn
data = load_boston()
```

```python
data.keys()
```

В этом словаре: \
data - независимые переменые (X) \
target - зависимая переменная (y) \
feature_names - подписи к независмым переменным \
DESCR - текстовое описание датасета

```python
# преобразуем данные в датафрейм
# в X передадим незаивсимые переменные 'data', имена колонок из 'feature_names'
X = pd.DataFrame(data['data'], columns=data['feature_names'])
```

```python
# посмотрим первые 5 строк
X.head()
```

```python
# описание датасета
print(data['DESCR'])
```

```python
# посмотрим размерность
X.shape
```

```python
# посмотрим базовые статистики
X.describe()
```

Факторов много, и они все разного порядка, от тысячных долей до сотен.
Для обучения модели линейной регресии нужно нормализовать данные (привести к одному порядку).
Нормализация признаков позволяет всем признакам вносить одинаковый вклад (или, что более точно, позволяет добавлять признаки в зависимости от их важности, а не их масштаба).
Используем для нормализации данных метод стандартизации: вычтем из данных среднее и поделим на стандартное отклонение.

```python
X = (X - X.mean()) / X.std()
```

```python
X.head()
```

```python
# запишем зависимую переменную в y
y = pd.Series(data['target'])
y.name = 'target'
y.head()
```

```python
# для удобства склеим X и y
z = pd.concat([X, y], axis=1)
z.head()
```

## Корреляция

```python
# визуализируем взаимосвязь преступности и цены дома
sns.scatterplot(data=z, x='CRIM', y='target')
```

Видно, что:
- с возрастанием криминальности цена на недвижимость падает
- есть странное поведение очень дорогой недвижимости в области с небольшой преступностью 

```python
# можем использовать pairplot для изучения взаимосвязи между множеством признаков
sns.pairplot(z, kind='reg')
```

```python
# изучаем взаимосвязь между населением низкого соц. статуса и количеством комнат в доме
sns.scatterplot(data=z, x='LSTAT', y='RM')
```

Видно, что есть обратная линейная зависимость. Чем больше доля людей с низким соц. статусом, тем дешевле жилье в этом районе.

```python
# изучаем взаимосвязь между ценой и количеством комнат в доме
sns.scatterplot(data=z, x='RM', y='target')
```

Также видна линейная зависимость - чем больше комнат, тем дороже жилье.

## Регрессия

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# делим все данные на два случайных подмножества, 
# 70% данных на обучение, 30% на тест
# random_state - суть этого параметра в воспроизводимых случайных значениях. 
# Т.е. если явно задать значение random_state отличным от None, то генерируемые псевдослучайные величины будут иметь одни и те же значения при каждом вызове.
# При этом само значение роли не играет, может быть и 42, и 0, и 21 и т.д.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

#### Постороим модель по признаку 'CRIM'

```python
# строим модель линейной регресии
model = LinearRegression()
# метод обучается на данных и подбирает оптимальные коэффициенты
model.fit(X_train[['CRIM']], y_train)  
```

```python
# коэф. a
model.coef_
```

```python
# коэф. b
model.intercept_
```

Получили такую модель: 

$y_{pred} = ax + b = -3,694x + 22,954$

```python
# метод предсказывает значение y на тестовых данных
y_pred = model.predict(X_test[['CRIM']])
y_pred
```

```python
# можно сверить предсказанные результаты с изначальными значениями
y_test
```

Далее нужно оценить адекватность модели, хорошо ли она предсказывает.
Способы для оценки:
- коэффициент детерминации 
- анализ остатков

#### Коэффициент детерминации: 


```python
# метод возвращает значение коэффициента детерминации
model.score(X_test[['CRIM']], y_test) 
```


В данном случае коэф. очень слабый = 0,13 \
Полученная модель плохо описывает зависимость цены на недвижимость от уровня криминальности. Скорее всего это связано со странным скоплением данных по дорогой недвижимости в  области с небольшой преступностью.

```python
# визуализируем полученную модель
sns.scatterplot(x=X_test.CRIM, y=y_test)
sns.lineplot(x=X_test.CRIM, y=y_pred, color='r')
```

#### Анализ остатков

Если модель подобрана правильно, то:
- остатки ведут себя достаточно хаотично
- в остатках нет систематической составляющей, резких выбросов
- в чередовании знаков нет закономерностей.



```python
# посмотрим на остатки, для этого из фактических значений вычтем предсказанные
y_resid = y_test- y_pred
sns.scatterplot(x=X_test['CRIM'], y=y_resid)
```

```python
# гистограмма распределения остатков
sns.histplot(y_resid)
```

## Альтернативный способ из statsmodel

```python
from statsmodels import api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
X_const = sm.add_constant(X_train) # техническая особенность библиотеки, надо руками добавить константу (intercept)
```

```python
X_const
```

```python
model = sm.OLS(y_train, X_const)
results = model.fit()
print(results.summary())
```

```python
print('Parameters: ', results.params)
print('R2: ', results.rsquared)
```

```python
prstd, iv_l, iv_u = wls_prediction_std(results) 

# для каждой точки получаем не просто сам прогноз, но и доверительный интервал
# с увеличением количества факторов, этот интервал будет уменьшаться
fig, ax = plt.subplots(figsize=(18, 8))

sns.lineplot(x=X_const.CRIM, y=y_train, marker='o', label="data")
sns.lineplot(x=X_const.CRIM, y=results.fittedvalues, color='r', linestyle='--', label="OLS")
sns.lineplot(x=X_const.CRIM, y=iv_u, color='g', linestyle='--')
sns.lineplot(x=X_const.CRIM, y=iv_l, color='g', linestyle='--')
plt.legend()
```
