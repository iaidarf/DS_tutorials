## Закон больших чисел
Согласно ЗБЧ: среднее значение большого числа независимых случаных вличин с одинаковым распределением стремится к мат.ожиданию.\
Проведем эксперимет с подбрасыванием кубика.\
Матожидание среднего значения всех очков:
1/6 * (1+2+3+4+5+6) = 3,5 \
При достаточно большом количестве бросков среднее значение всех очков будет стремиться к 3,5.


```python
import numpy as np
import seaborn as sns
```

```python
n = 1000 # кол-во испытаний
x = [] # массив для записи выпавших значений кубика
means = [] # среднее значение очков
for _ in range(n):
    x.append(np.random.randint(1, 7))
    means.append(np.array(x).mean())
```

```python
line = 3.5
graph = sns.lineplot(data = means)
graph.axhline(line, color = 'r')

```
